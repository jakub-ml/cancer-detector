```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier,VotingClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df=pd.read_csv('data.csv')
df.head()
```

```python
df.describe()
```

```python
print('Data shape: {}'.format(df.shape))
```

```python
df.isna().sum()
```

```python
y=df[['diagnosis']]
df=df.drop(columns=['Unnamed: 32','diagnosis','id'])
df
```

```python
corr=df.corr()
a=np.array(corr.columns)
lista=a[corr.sum()>10]
corr=corr[corr.sum()<10].drop(columns=lista)
sns.heatmap(corr)
```

### Preprocessing 1 (None)

```python
# No standarization or PCA
X = df
y = y['diagnosis']
```

### Preprocessing 2 (Standarize)

```python
# Standarize
scaler=StandardScaler()
std_df = pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
std_df.head()
X = df
y = y['diagnosis']
```

### Preprocessing 3 (Standarize + PCA)

```python
pca = PCA(n_components=0.95)
scaler = StandardScaler()

std_df = pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
pca_df = pca.fit_transform(std_df)
print(pca.explained_variance_ratio_)

X = pca_df
y = y['diagnosis']
```

## Trenowanie modelu

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
print('Wielkość danych treningowych {} i testowych {}'.format(X_train.shape[0], X_test.shape[0]))
```

#### KNN

```python
# single model training 
model = KNeighborsClassifier(n_neighbors = 70)
model.fit(X_train, y_train)

y_train_data = model.predict(X_train)
y_pred = model.predict(X_test)

acc_train = accuracy_score(y_train, y_train_data)
acc_pred = accuracy_score(y_pred, y_test)

print('Train accuracy: {}'.format(acc_train))
print('Test accuracy: {}'.format(acc_pred))
```

```python
# n_neighbors search
kneighbors=np.arange(1,101,1)
score=[]
for n in kneighbors:
    model=KNeighborsClassifier(n_neighbors=n)
    model.fit(X_train, y_train)
    y_pred=model.predict(X_test)
    score.append(accuracy_score(y_pred, y_test))
plt.plot(kneighbors,score,'--')
plt.ylabel('Liczba sąsiadów')
plt.xlabel('Wynik modelu')
plt.title('Wykres zależności liczby sąsiadów od otrzymywanego wyniku')
plt.grid()
plt.show()
print('Jak widać na otrzymanym wykresie, {} sąsiadów jest liczbą o najlepszej skuteczności modelu w wysokości: {}%'.format(kneighbors[score.index(np.array(score).max())],(100*np.array(score).max())))
print('W ten sposób poprawiliśmy skuteczność modelu o {}%'.format((100*(np.array(score).max()-acc_pred))))
```

```python
# Grid Search with Kfold
model = KNeighborsClassifier()
params = {'n_neighbors':np.arange(1,101,1), 'leaf_size':np.array([4, 8, 16, 32])}
kf = KFold(n_splits = 3, shuffle=True, random_state = 126)
model_cv = GridSearchCV(model, params, cv = kf)
model_cv.fit(X_train, y_train)
print('Best parameters {}, with accuracy {}'.format(model_cv.best_params_, model_cv.best_score_))
```

```python
# other models comparison
modele={'KNeighborsClassifier':KNeighborsClassifier(),'LogisticRegression':LogisticRegression(max_iter=500),'SVC':SVC(),'DecisionTreeClassifier':DecisionTreeClassifier(),'Ada':AdaBoostClassifier(),'Gradien':GradientBoostingClassifier()}
for name,model in modele.items():
    model=model
    model.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    print('Wynik dla modelu {} wynosi: {}'.format(name,100*(accuracy_score(y_pred,y_test))))
```

```python
# selecting 3 best models to build voiting classifier
best={'KNeighborsClassifier' : [KNeighborsClassifier(), {'n_neighbors':np.arange(1,101,1), 'leaf_size':np.array([4, 8, 16, 32])}], 'Ada':[AdaBoostClassifier(),{'n_estimators':np.arange(2, 60, 2)}],'Gradien':[GradientBoostingClassifier(),{'max_depth':np.array([1, 3, 6, 10]), 'min_samples_leaf': np.array([1, 2, 3, 4]), 'n_estimators':np.array([10, 40, 70, 100])}]}

for name, _ in best.items():
    print(name)
```

```python
modele = []
for name, model_dict in best.items():
    model = model_dict[0]
    kf = KFold(n_splits = 3, shuffle=True, random_state = 126)
    params = model_dict[1]
    model_cv = GridSearchCV(model, params, cv = kf)
    model_cv.fit(X_train, y_train)
    modele.append((name, model_cv))
    print(f"Model {name} ready.")
```

```python
modele_score = []
for elem in modele:
    y_pred = elem[1].predict(X_test)
    modele_score.append(accuracy_score(y_pred, y_test))
    print('Model {} accuracy: {}'.format(modele.index(elem)+1, accuracy_score(y_pred, y_test)))
```

```python
# voting classifier score
voting=VotingClassifier(estimators=modele)
voting.fit(X_train,y_train)
y_pred=voting.predict(X_test)
score=accuracy_score(y_test, y_pred)
score
```

```python
x=np.array(['KNeighborsClassifier','AdaBoostClassifier','GradientBoostingClassifier','LogisticRegression','VotingClassifier'])
y=np.array([model_cv.best_score_, modele_score[0], modele_score[1], modele_score[2], score])
plt.bar(x, y*100)
plt.title('Wyniki poszczególnych modeli')
plt.xlabel('Modele')
plt.ylabel('Wyniki')
plt.grid()
plt.ylim(92, 100)
plt.xticks(rotation = 45)
```
